```python
count = 0
class Clock12:

    @property
    def max_limit(self):
        return 100

    def __init__(self, *args):
        self.number = count
        self.contrafact=False
        if self.number > self.max_limit:
            self.contrafact=True
        self._h: int=0
        self._m: int=0
        self.increase()
        if len(args) == 0:
            return
        elif len(args)>2:
            raise TypeError('Clock 12 can`t take more than 2 positional arguments!')
        elif len(args) == 2:
            self.set_time(args[0], args[1])
        else:
            if isinstance(args[0], int):
                self._from_int(args[0])
            elif isinstance(args[0], Clock12):
                self._assign(args[0])
            else:
                raise TypeError('The single argument of Clock12.__init__ should be int or Clock12')


    @property
    def hour(self):
        return self._h

    @property
    def minute(self):
        return self._m

    @property
    def time(self):
        return self.hour * self.max_minute() + self.minute

    @property
    def step(self):
        return 1

    @time.setter
    def time(self, minute: int):
        self._from_int(minute)

    def get_hour(self):
        return self._h

    def get_minute(self):
        return self._m

    def set_time(self,h: int, m: int):
        self._is_good_time(h, m)
        self._set_time(h, m)

    @classmethod
    def _is_good_time(cls, h, m):
        return cls._is_good_hour(h) and cls._is_good_minute(m)

    @classmethod
    def _is_good_hour(cls, h):
        if not isinstance(h, int):
            raise TypeError(str(type(h))+' is improper value for hour')
        if 0<=h<cls.max_hour():
            return True
        raise ValueError(str(h)+' is improper value for hour')

    @classmethod
    def _is_good_minute(cls, m):
        if not isinstance(m, int):
            raise TypeError(str(type(m)) + ' is improper value for minute')
        if 0 <= m < cls.max_minute():
            return True
        raise ValueError(str(m) + ' is improper value for minute')

    def _set_time(self, h, m):
        self._h=h
        self._m=m

    def _from_int(self, minute):
        minute %= 60*12
        self._set_time(*divmod(minute, 60))

    def _assign(self, other):
        self._set_time(other.get_hour(), other.get_minute())

    def input(self):
        h0=int(input('Enter hour value (0...11): '))
        m0 = int(input('Enter minute value (0...59): '))
        self.set_time(h0, m0)

    def output(self):
        print(f'h = {self.get_hour()}, m ={ self.get_minute()}')

    def _is_equal(self, other):
        return self.get_hour()==other.get_hour() and self.get_minute()==other.get_minute( )

    def copy(self):
        return self.__class__(self)

    def is_contrafact(self):
        return self.contrafact

    @staticmethod
    def max_hour():
        return 12

    @staticmethod
    def max_minute():
        return 60

    def increase(self):
        global count
        count += self.step

    @classmethod
    def first_pair(cls):
        return cls(8, 40)





c=Clock12(159)
c1=Clock12(c)
c2=Clock12()
c3=c.copy()
c.output()
c1.output()
c2.output()
c3.output()
print(c._is_equal(c3))
print(c._is_equal(c1))
c4=c.first_pair()
c4.output()
print(c4.minute)
c7=Clock12(7, 4)
print(c7.is_contrafact(), c7.number)
c8=Clock12()
print(c8.is_contrafact(), c8.number)


```
