1. Задано множину s, що складається з цілих чисел. Записати ВИРАЗ, значенням якого є множина, що складається з елементів множини s, що діляться на 13.

2. Записати ВИРАЗ, значенням якого є множина, що складається з квадратних коренів цілих чисел від 15 до 2001 (включно), що не діляться на 2.

3. Задано множину s, що складається з цілих чисел. Записати ВИРАЗ, значенням якого є множина, що складається з з непарних елементів множини s.

```python
s={6, 7, 4, 0, 6, 13, 52, 4, 26}
```

```python
a={el for el in s if not el%13}
a
```

```python
import math
b={math.sqrt(el) for el in range(15, 2002) if el%2}
b
```

1. Записати ВИРАЗ, значенням якого є словник, ключами якого є цілі числа від 14 до 2012 (включно), що діляться на 7, а ключам відповідають їх квадратні корені

2. Записати ВИРАЗ, значенням якого є словник, ключами якого є цілі числа від 12 до 2000 (включно), що не діляться на 3, а ключам відповідають їх квадратні корені

3. Записати ВИРАЗ, значенням якого є словник, ключами якого є цілі числа від 11 до 2012 (включно), що діляться на 5, а ключам відповідають їх куби. 

4. Нехай змінна s позначає послідовність цілих, по якій можна ітерувати. Написати вираз, що перевіряє, чи правда, що всі елементи послідовності більші за задане число num.

```python
d={el:math.sqrt(el) for el in range(14, 2013) if not el%7}
d
```

```python
e={el:math.sqrt(el) for el in range(12, 2001) if el%3}
e
```

```python
f={el:el**3 for el in range(11, 2013) if not el%5}
f
```
